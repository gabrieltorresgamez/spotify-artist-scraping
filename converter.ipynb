```python
# GET ALL THE ARTISTS FROM data/artists_info/X.json
artists_info_path = "data/artists_info"
files = os.listdir(artists_info_path)
files = [artists_info_path + "/" + file for file in files if file.endswith(".json")]
files

artists = pd.DataFrame()
for file in files:
    with open(file) as f:
        data = json.load(f)

    temp = pd.DataFrame({"artist_id": data['id'], "artist": data['name'], "followers": data['followers']['total'], "popularity": data['popularity']}, index=[0])
    artists = pd.concat([artists, temp])

artists = artists.reset_index(drop=True)
artists = artists.drop_duplicates()
artists

os.makedirs("export", exist_ok=True)
artists.to_csv("export/artists.csv", index=False)
```

```python
tracks_path = "data/artists_discography"
artists = os.listdir(tracks_path)
artists

tracks = pd.DataFrame()
for artist in artists:
    for track in os.listdir(tracks_path + "/" + artist):
        with open(tracks_path + "/" + artist + "/" + track) as f:
            data = json.load(f)

        temp = pd.DataFrame({"track_id": data['id'], "track_name": data['name'], "popularity": data['popularity'], "explicit": data['explicit']}, index=[0])
        temp = temp.rename(columns={"id": "track_id"})
        tracks = pd.concat([tracks, temp])

tracks = tracks.reset_index(drop=True)
tracks = tracks.drop_duplicates()
tracks
```

```python
tracks_audio_features_path = "data/tracks_audio_features"
tracks_ = os.listdir(tracks_audio_features_path)
tracks_

audio_features = pd.DataFrame()
for track in tracks_:
    with open(tracks_audio_features_path + "/" + track) as f:
        data = json.load(f)

    temp = pd.DataFrame(data, index=[0])
    temp = temp.drop(columns=['type', 'uri', 'track_href', 'analysis_url'])
    temp = temp.rename(columns={"id": "track_id"})
    audio_features = pd.concat([audio_features, temp])

audio_features = audio_features.reset_index(drop=True)
audio_features = audio_features.drop_duplicates()
audio_features
```

```python
tracks = pd.merge(tracks, audio_features, on="track_id", how="outer")
tracks = tracks.drop_duplicates()
tracks.to_csv("export/tracks.csv", index=False)
```

```python
artist_tracks = pd.DataFrame()
for artist in artists:
    for track in os.listdir(tracks_path + "/" + artist):
        with open(tracks_path + "/" + artist + "/" + track) as f:
            data = json.load(f)

        temp = pd.DataFrame({"track_id": data['id'], "artist_id": [artist['id'] for artist in data['artists']]})
        artist_tracks = pd.concat([artist_tracks, temp])

artist_tracks = artist_tracks.reset_index(drop=True)
artist_tracks = artist_tracks.drop_duplicates()
artist_tracks.to_csv("export/artists_tracks.csv", index=False)
```
